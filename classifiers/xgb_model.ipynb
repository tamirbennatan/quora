### 1.0 Setting up testing environment

#### 1.1 Keeping features that would like to experiment with.

Some of these features don't make sense in the context of boosted trees (e.g word embeddings), or are intuitively meaningless (e.g partial token set ratio with stopwords). There are also pairs of features which don't make to keep (e.g num characters in q1, and number of characters in q1 excluding stopwords). 

Further, there are pairs of features that would clearly be redundunt if included. For example, partial token sort ratio and paritial token sort ratio without stopwords do not both need to be incldued. 

Boosted trees, like other ensemble tree-based methods, is relatively robust with regard to correlated features. Thus, it's not of huge concern to include features that are likely correlated. The only reason why I exclude some features it to speed up training, and hopefully keep the the feature importance weights (f scores)

### 3. Approximate the number and depth of the estimators

There are a bunch of parameters to tune when using `XGboost`, but the one which affects the time it takes to train are the number of estimators (trees) and the depth of each tree. 

Further, these parameters tend to yield a point of diminishing return rather quicly. Thus, I'll get a rough approximation as to the optimal number of trees and the depth of trees using a `GridSearchCV`, and then move on to more fine parameter tuning once I have a better idea of the scale of the size of my ensemble


```python
# a vanilla XGboost model
model1 = XGBClassifier(random_state = 550)
```

```python
# create a parameter grid for the number of trees and depth of trees
n_estimators = [50, 100, 150, 200]
max_depth = [2, 4, 6, 8]
param_grid = dict(max_depth=max_depth, n_estimators=n_estimators)
```

```python
# Create gridsearch cross validation object

grid_search = GridSearchCV(model1, param_grid, scoring='neg_log_loss', n_jobs=-1, cv=6, verbose=1)
```

```python
%time grid_result = grid_search.fit(X_train, y_train)
#print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
grid_result.grid_scores_
```

```python
grid_result.best_params_
```

```python
grid_result.best_score_
```

We can see that the algorithm performs best with many deep trees. THe fact that the log loss reduces subtancially from 200 trees to 150 trees with a depth of 8 makes me think that I unerestimated the number of estimators needed. A more appropriate range should be [200, 400, ... 1000].

### 7. Abridged Features

In the selected model, I've included over two dozen features. Surely, some do not provide much explanitory power, and only make the model unwieldly to train and interpret.

Using a simple feature important score called the _f-score_ (which counts the number of times a tree was split on a feature) I can drop the least important features. This will help train future models faster, and make the marginal gain of newly added features more clear. 

#### 7.0 Loading the baseline XBG model

```python
xgb_baseline = pkl.load(open("../models/xgboost_model1.pickle.dat", "rb"))
```

```python
feature_importance(xgb_baseline)
```

The importance of the features are skewed. Further, some of the features are very similar. 

For example, the features `wratio_sw` and `wratio` are moth variants of Levenshtein distance scores applied to the questions - the only difference is that one removes stopwords and the other doesn't. 

It appears that these measures are not important from the chart - but this is misleading. Because some features are highly correlated (see below), it may seem like a feature is unimportant, but it is really that a tree is indifferent between splitting on that feature or another, because they are highly correlated. 

Thus, it is helpful to remove redundant featuers for interpretability purposes. 

#### 7.1 Feature Correlation

```python
# correlation of feature values
corr = X_train.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

#### 7.2 Abridged Features

I'll remove the following features, for the following reasons:

- wratio: The wratio is a linear combination of the other Levenshtein distance features. This is why it is so highly correlated with so many other features. As such, it doesn't make sense to keep features for t both the wratio and the wratio, when stopwords are excluded. 
- len_intersection_sw, num_words_diff, and num_words_diff_sw: These are some of the least important features (according to the f-score). They are also highly correlated with more imporant features, like the number of characters in each sentence. As such, any the variance they might explain is likely captured by the other variables. 
- partial_token_set_ratio_sw and jaccaard_distance_sentence_embeddings: These features have very few univariate variability, and so they do not provide any information. 

```python
# abridged featrues
abridged_features = [ft for ft in training_features if \
                    ft not in [u'wratio', u'len_intersection_sw',u'num_words_diff', u'num_words_diff_sw',
                                 u'partial_token_set_ratio_sw', u'jaccard_distance_sentence_embeddings' ]]
```
